```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Carrega os DataFrames a partir dos arquivos CSV
plan_base = pd.read_csv('/content/items_titles.csv')
plan_test = pd.read_csv('/content/items_titles_test.csv')

# Certifique-se de que as colunas 'ITE_ITEM_TITLE' não contenham valores nulos
plan_base = plan_base.dropna(subset=['ITE_ITEM_TITLE'])
plan_test = plan_test.dropna(subset=['ITE_ITEM_TITLE'])

# Obtém os valores da coluna 'ITE_ITEM_TITLE' como listas
texto_a_ser_comparado = plan_test['ITE_ITEM_TITLE'].values
texto_fonte = plan_base['ITE_ITEM_TITLE'].values

# Pré-processamento de texto
def preprocess_text(text):
    # Converte para minúsculas
    text = text.lower()
    # Remove caracteres especiais, pontuações, etc.
    text = ''.join(e for e in text if e.isalnum() or e.isspace())
    return text

# Aplica o pré-processamento aos textos
texto_a_ser_comparado = [preprocess_text(text) for text in texto_a_ser_comparado]
texto_fonte = [preprocess_text(text) for text in texto_fonte]

# Numero de n_gramas
n = 1

# Instancia o contador de ngramas
counts = CountVectorizer(analyzer='word', ngram_range=(n, n))

# Cria um dicionário de n-gramas
vocab2int = counts.fit(texto_a_ser_comparado + texto_fonte).vocabulary_

# Printa dicionário de palavras: index
print(vocab2int)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from unidecode import unidecode

# Função de pré-processamento
def preprocess_text(text):
    text = text.lower()
    text = unidecode(''.join(e for e in text if e.isalnum() or e.isspace()))
    return text

# Carrega os DataFrames a partir dos arquivos CSV
plan_base = pd.read_csv('/content/items_titles.csv')
plan_test = pd.read_csv('/content/items_titles_test.csv')

# Certifique-se de que as colunas 'ITE_ITEM_TITLE' não contenham valores nulos
plan_base = plan_base.dropna(subset=['ITE_ITEM_TITLE'])
plan_test = plan_test.dropna(subset=['ITE_ITEM_TITLE'])

# Tira os acentos


# Obtém os valores da coluna 'ITE_ITEM_TITLE' como listas
texto_a_ser_comparado = plan_test['ITE_ITEM_TITLE'].values
texto_fonte = plan_base['ITE_ITEM_TITLE'].values

# Aplica o pré-processamento aos textos
texto_a_ser_comparado = [preprocess_text(text) for text in texto_a_ser_comparado]
texto_fonte = [preprocess_text(text) for text in texto_fonte]

# Numero de n_gramas
n = 1

# Instancia o contador de ngramas
counts = CountVectorizer(analyzer='word', ngram_range=(n, n))

# Transforma os textos em vetores de contagem
vetores_a_ser_comparado = counts.fit_transform(texto_a_ser_comparado)
vetores_fonte = counts.transform(texto_fonte)

# Calcula a similaridade de cosseno entre os vetores
similaridade_cosseno = cosine_similarity(vetores_a_ser_comparado, vetores_fonte)

# Cria um DataFrame com os textos e a similaridade
df_similaridade = pd.DataFrame(similaridade_cosseno, columns=plan_base['ITE_ITEM_TITLE'], index=plan_test['ITE_ITEM_TITLE'])

# Adiciona uma coluna com a média da similaridade para cada texto de teste
df_similaridade['Similaridade Média'] = similaridade_cosseno.mean(axis=1)

# Exibe o DataFrame
print(df_similaridade)

# Salva o DataFrame em um arquivo CSV
df_similaridade.to_csv('/content/similaridade_resultados.csv', index=False)
```
